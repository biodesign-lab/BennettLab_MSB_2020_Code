### Figure S2

This code creates Figure S2.

```python
import numpy as np
import seaborn as sns
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
# Load the data from matlab .mat file
circuits_data = sio.loadmat('all_circuits_12_03_2018.mat')
numDataPoints=len(circuits_data['all_od'][:,0,0,0])
N = 49
time_vec=np.array(range(0,N*10,10))
day=0 #day 1 of the experiments
ts=270 #start of the saturation pahse of the cell growth 

#Store the data for the 12 replicates of each experiment
X_all_cascade=np.zeros([N, 12])
Y_all_cascade=np.zeros([N, 12])
Z_all_cascade=np.zeros([N, 12])
Exper=0 #cascade 
for repl in range(12):
    X=circuits_data['all_flu'][0:N,repl,Exper,0,day]
    Y=circuits_data['all_flu'][0:N,repl,Exper,1,day]
    Z=circuits_data['all_flu'][0:N,repl,Exper,2,day]
    X_all_cascade[:, repl]=X-np.mean(X[0:11])
    Y_all_cascade[:, repl]=Y-np.mean(Y[0:11])
    Z_all_cascade[:, repl]=Z-np.mean(Z[0:11])

X_all_inverter=np.zeros([N, 12])
Y_all_inverter=np.zeros([N, 12])
Z_all_inverter=np.zeros([N, 12])
Exper=1 #cascade 
for repl in range(12):
    X=circuits_data['all_flu'][0:N,repl,Exper,0,day]
    Y=circuits_data['all_flu'][0:N,repl,Exper,1,day]
    Z=circuits_data['all_flu'][0:N,repl,Exper,2,day]
    X_all_inverter[:, repl]=X-np.mean(X[0:11])
    Y_all_inverter[:, repl]=Y-np.mean(Y[0:11])
    Z_all_inverter[:, repl]=Z-np.mean(Z[0:11])    

X_all_fanout=np.zeros([N, 12])
Y_all_fanout=np.zeros([N, 12])
Z_all_fanout=np.zeros([N, 12])
Exper=4 #cascade 
for repl in range(12):
    X=circuits_data['all_flu'][0:N,repl,Exper,0,day]
    Y=circuits_data['all_flu'][0:N,repl,Exper,1,day]
    Z=circuits_data['all_flu'][0:N,repl,Exper,2,day]
    X_all_fanout[:, repl]=X-np.mean(X[0:11])
    Y_all_fanout[:, repl]=Y-np.mean(Y[0:11])
    Z_all_fanout[:, repl]=Z-np.mean(Z[0:11]) 
    
    
sns.set()
sns.set(font_scale=1.2)
sns.set_style('ticks')
red_colors = sns.color_palette("Reds_d",9)
blue_colors = sns.color_palette("Blues_d",9)
orange_colors = sns.color_palette("YlOrBr_d",9)
Fig2 = plt.figure(figsize=(12,10))

#Plot 12 replicates of strain X for cascade
plt.subplot(3,3,1)
plt.plot(time_vec,X_all_cascade, color=red_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(X_all_cascade, axis=1), color=red_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,1500),(0,1500))
plt.title('Cascade', fontsize=20)
sns.despine()


#Plot 12 replicates of strain Y for cascade
plt.subplot(3,3,4)
plt.plot(time_vec,Y_all_cascade, color=blue_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(Y_all_cascade, axis=1), color=blue_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,15000),(0,15000))
sns.despine()

#Plot 12 replicates of strain Z for cascade
plt.subplot(3,3,7)
plt.plot(time_vec,Z_all_cascade, color=orange_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(Z_all_cascade, axis=1), color=orange_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,50000),(0,50000))
sns.despine()



#Plot 12 replicates of strain X for inverter circuit
plt.subplot(3,3,2)
plt.plot(time_vec,X_all_inverter, color=red_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(X_all_inverter, axis=1), color=red_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,2500),(0,2500))
plt.title('Inverter', fontsize=20)
sns.despine()


#Plot 12 replicates of strain Y for inverter
plt.subplot(3,3,5)
plt.plot(time_vec,Y_all_inverter, color=blue_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(Y_all_inverter, axis=1), color=blue_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,25000),(0,25000))
sns.despine()

#Plot 12 replicates of strain Z for inverter
plt.subplot(3,3,8)
plt.plot(time_vec,Z_all_inverter, color=orange_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(Z_all_inverter, axis=1), color=orange_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,10000),(0,10000))
sns.despine()





#Plot 12 replicates of strain X for fanout circuit
plt.subplot(3,3,3)
plt.plot(time_vec,X_all_fanout, color=red_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(X_all_fanout, axis=1), color=red_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,1500),(0,1500))
plt.title('Fanout', fontsize=20)
sns.despine()


#Plot 12 replicates of strain Y for fanout
plt.subplot(3,3,6)
plt.plot(time_vec,Y_all_fanout, color=blue_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(Y_all_fanout, axis=1), color=blue_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,15000),(0,15000))
sns.despine()

#Plot 12 replicates of strain Z for fanout
plt.subplot(3,3,9)
plt.plot(time_vec,Z_all_fanout, color=orange_colors[4], lw=0.5, alpha=0.7)   
plt.plot(time_vec, np.mean(Z_all_fanout, axis=1), color=orange_colors[4], lw=2)
plt.axvline(ts, linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,240,480),(0,240,480))
plt.yticks((0,8000),(0,8000))
sns.despine()





Fig2.subplots_adjust(wspace=0.4)


Fig2.savefig("FigS2_otherNetworks.pdf", bbox_inches='tight', transparent=True)
```
