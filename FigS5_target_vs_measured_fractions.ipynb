### Figure S5

Here we create Figure S5 in the supplement.

```python
import numpy as np
import seaborn as sns
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
# Load the data from matlab .mat file
measured_fracs = sio.loadmat('measured_fracs_time0.mat')
target_fracs = sio.loadmat('target_fracs.mat')


# Get the measured and target fractions
exper = 0 #r_Z=0.33
day=0 #day 1 of the fixZ experiment
rX_m_d1 = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d1 = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d1 = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z
day=1 #day 2 of the fixZ experiment
rX_m_d2 = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d2 = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d2 = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z
day=2 #day 3 of the fixZ experiment
rX_m_d3 = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d3 = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d3 = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z
rX_target=target_fracs['ideal_fracs'][0,:,exper]
rY_target=target_fracs['ideal_fracs'][1,:,exper]
rZ_target=target_fracs['ideal_fracs'][2,:,exper]

exper = 1 #r_X=0.2
day=0 #day 1 of the fixZ experiment
rX_m_d1_fixX = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d1_fixX = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d1_fixX = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z
day=1 #day 2 of the fixZ experiment
rX_m_d2_fixX = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d2_fixX = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d2_fixX = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z
day=2 #day 3 of the fixZ experiment
rX_m_d3_fixX = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d3_fixX = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d3_fixX = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z
rX_target_fixX=target_fracs['ideal_fracs'][0,:,exper]
rY_target_fixX=target_fracs['ideal_fracs'][1,:,exper]
rZ_target_fixX=target_fracs['ideal_fracs'][2,:,exper]



sns.set()
sns.set(font_scale=1.2)
sns.set_style('ticks')
red_colors = sns.color_palette("Reds_d",9)
blue_colors = sns.color_palette("Blues_d",9)
orange_colors = sns.color_palette("YlOrBr_d",9)
strains = ['X', 'Y', 'Z']
colors = [red_colors[5], blue_colors[5], orange_colors[5]]

Fig2 = plt.figure(figsize=(8,8))


# Plot the measured fractions against target ones in the first experiment 
# Plot r_X
strain = 0
plt.subplot(3,2,1)
plt.plot(rX_target,rX_m_d1, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot(rX_target,rX_m_d2, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot(rX_target,rX_m_d3, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot([0, 1], [0, 1], linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,0.5,1),(0,0.5,1))
plt.yticks((0,0.5,1),(0,0.5,1))
plt.title('Experiment with $r_Z=0.33$', fontsize=18)
plt.xlabel('target $r_X$', fontsize=16)
plt.ylabel('measured $r_X$', fontsize=16)
sns.despine()


# Plot r_Y
strain = 1
plt.subplot(3,2,3)
plt.plot(rY_target,rY_m_d1, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot(rY_target,rY_m_d2, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot(rY_target,rY_m_d3, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot([0, 1], [0, 1], linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,0.5,1),(0,0.5,1))
plt.yticks((0,0.5,1),(0,0.5,1))
plt.xlabel('target $r_Y$', fontsize=16)
plt.ylabel('measured $r_Y$', fontsize=16)
sns.despine()


# Plot r_Z
strain = 2
plt.subplot(3,2,5)
measured_rZ_vals=np.concatenate((rZ_m_d1, rZ_m_d2, rZ_m_d3))
plt.hist(measured_rZ_vals, 7, label='measured values', color=colors[strain])
plt.axvline(0.33, linestyle='--', color='k', lw=2, alpha=0.9, label='target $r_Z$')
plt.xlabel('$r_Z$', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.legend(fontsize=14, frameon=False, bbox_to_anchor=(0.6, .5), handletextpad=0.5)
plt.xlim([0.2, 0.6])
sns.despine()


# Plot the measured fractions against target ones in the second experiment (r_X=0.2)
# Plot r_X
strain = 0
plt.subplot(3,2,2)
measured_rX_vals=np.concatenate((rX_m_d1_fixX, rX_m_d2_fixX, rX_m_d3_fixX))
plt.hist(measured_rX_vals, 4, label='measured values', color= colors[strain])
plt.axvline(0.2, linestyle='--', color='k', lw=2, alpha=0.9, label='target $r_X$')
plt.title('Experiment with $r_X=0.2$', fontsize=18)
plt.xlabel('$r_X$', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.legend(fontsize=14, frameon=False, bbox_to_anchor=(0.5, 1), handletextpad=0.5)
plt.xlim([0, 0.4])
sns.despine()

# Plot r_Y
strain = 1
plt.subplot(3,2,4)
plt.plot(rY_target_fixX,rY_m_d1_fixX, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot(rY_target_fixX,rY_m_d2_fixX, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot(rY_target_fixX,rY_m_d3_fixX, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot([0, 1], [0, 1], linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,0.5,1),(0,0.5,1))
plt.yticks((0,0.5,1),(0,0.5,1))
plt.xlabel('target $r_Y$', fontsize=16)
plt.ylabel('measured $r_Y$', fontsize=16)
sns.despine()

# Plot r_Z
strain = 2
plt.subplot(3,2,6)
plt.plot(rZ_target_fixX,rZ_m_d1_fixX, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot(rZ_target_fixX,rZ_m_d2_fixX, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot(rZ_target_fixX,rZ_m_d3_fixX, linestyle='none', marker='.', markersize=7, color=colors[strain])   
plt.plot([0, 1], [0, 1], linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,0.5,1),(0,0.5,1))
plt.yticks((0,0.5,1),(0,0.5,1))
plt.xlabel('target $r_Z$', fontsize=16)
plt.ylabel('measured $r_Z$', fontsize=16)
sns.despine()

Fig2.subplots_adjust(wspace=1.3, hspace=0.5)

#Export figure
Fig2.savefig("FigS5_target_vs_measured_fracs.pdf", bbox_inches='tight', transparent=True)
```
